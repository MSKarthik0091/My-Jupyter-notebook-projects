```python
# Taking the first 10 classes as working dataframe
working_df = df[df['landmark_label'].isin(sorted(df['landmark_label'].unique())[:10])]
print(working_df)
working_df['landmark_label'].value_counts()
```

```python
# Shuffling the working dataframe thrice for better randomness
working_df = working_df.sample(frac=1, random_state=42).reset_index(drop=True)
working_df = working_df.sample(frac=1, random_state=42).reset_index(drop=True)
working_df = working_df.sample(frac=1, random_state=42).reset_index(drop=True)

print(working_df)
```

```python
from sklearn.model_selection import train_test_split

# Split into train and val
train_df, val_df = train_test_split(working_df, test_size=0.2, stratify = working_df['landmark_label'], random_state=42)
```

```python
print(len(train_df))
print(len(val_df))
working_num_classes = working_df['landmark_label'].nunique()
print(working_num_classes)
```

```python
print(train_df)
```

```python
from tensorflow.keras import layers, models, optimizers
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import pandas as pd
import os

# Image dimensions (you can tune this)
IMG_SIZE = (224, 224)
BATCH_SIZE = 64 # As per your data

# --- DATASET LOADING ---
train_df['landmark_label'] = train_df['landmark_label'].astype(str)
val_df['landmark_label'] = val_df['landmark_label'].astype(str)

# Image generators
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=10,
    width_shift_range=0.1,
    height_shift_range=0.1,
    zoom_range=0.2,
    horizontal_flip=True
)
val_datagen = ImageDataGenerator(rescale=1./255)

train_gen = train_datagen.flow_from_dataframe(
    train_df,
    x_col='filepath',
    y_col='landmark_label',
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical'
)
val_gen = val_datagen.flow_from_dataframe(
    val_df,
    x_col='filepath',
    y_col='landmark_label',
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical'
)
```

```python
from tensorflow.keras.applications import EfficientNetB0
from tensorflow.keras.layers import GlobalAveragePooling2D, Dense, Dropout
from tensorflow.keras.models import Model
from tensorflow.keras import Input
from tensorflow.keras.optimizers import Adam

def build_efficientnetb0(input_shape=(224, 224, 3), num_classes=working_num_classes):
    # Load EfficientNetB0 backbone, pretrained on ImageNet
    backbone = EfficientNetB0(
        include_top=False,
        weights='imagenet',
        input_shape=input_shape
    )
    
    # Build head
    x = backbone.output
    x = GlobalAveragePooling2D()(x)
    x = Dropout(0.3)(x)
    x = Dense(512, activation='relu')(x)
    x = Dropout(0.3)(x)
    outputs = Dense(num_classes, activation='softmax')(x)
    
    return Model(inputs=backbone.input, outputs=outputs)

# Instantiate
model = build_efficientnetb0()

model.compile(
    optimizer=Adam(learning_rate=1e-4),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)


model.summary()
```

```python
# --- TRAINING --- for the first 10 landmarks only (10-class classification)
history = model.fit(
    train_gen,
    validation_data=val_gen,
    epochs=25,
    steps_per_epoch=len(train_gen),
    validation_steps=len(val_gen),
    verbose=1
)
```

```python
# Evaluate on validation data for the first 10 landmarks only (10-class classification)
loss, accuracy = model.evaluate(val_gen)
print(f"Validation Accuracy: {accuracy * 100:.2f}%\nValidation Loss: {loss:.5f}")
```

```python
# To save the entire model (architecture + weights + optimizer)
model.save("my_landmark_model_first_10_landmarks.h5")
```
